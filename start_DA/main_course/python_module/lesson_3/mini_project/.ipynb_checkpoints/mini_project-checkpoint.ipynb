```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

user_data = pd.read_csv('3_user_data.csv')
logs = pd.read_csv('3_logs.csv')
user_data = user_data.merge(logs, how='inner', on='client')
user_data
```

```python
print(f'Размер таблицы = {user_data.shape}\n')
print(f'Типы данных в таблице:\n{user_data.dtypes}\n')
print(f'Описательная статистика таблицы: \n{user_data.describe()}\n')
print(f'Число уникальных значений переменной platform = {user_data['platform'].nunique()}')
```

```python
client_operations = user_data \
    .query('success == True') \
    .groupby('client') \
    .agg({'client': 'count'}) \
    .rename(columns={'client': 'count'}) \
    .sort_values('count', ascending=False) \
    .reset_index() \
    .head(10)
client_operations
```

```python
max_op_clients_list = []
for i in range(9):
    max_op_clients_list.append(client_operations['client'][i])
max_op_clients_list = sorted(max_op_clients_list)
max_op_clients_list
```

```python
platform_operations = user_data \
    .query('success == True') \
    .groupby('platform') \
    .agg({'platform': 'count'}) \
    .rename(columns={'platform': 'count'}) \
    .sort_values('count', ascending=False) \
    .reset_index()
platform_operations
```

```python
max_op_platform_list = []
for i in range(platform_operations['platform'].shape[0]):
    max_op_platform_list.append(platform_operations['platform'][i])
max_op_platform_list
```

```python
premium_popularity = user_data \
    .query('premium == True') \
    .groupby('platform') \
    .agg({'platform': 'count'}) \
    .rename(columns={'platform': 'count'}) \
    .sort_values('count', ascending=False) \
    .reset_index()
premium_popularity
```

```python
sns.barplot(x='platform', y='count', hue='platform', data=premium_popularity)
```

```python
premium_age = user_data \
    .query('premium == True')
freemium_age = user_data \
    .query('premium == False')
freemium_age
```

```python
sns.displot(premium_age['age'], kde=True, alpha=.3, stat='density')
sns.displot(freemium_age['age'], kde=True, alpha=.3, stat='density')
```

```python
client_succ_operations = user_data \
    .query('success == True') \
    .groupby('client') \
    .agg({'client': 'count'}) \
    .rename(columns={'client': 'count'}) \
    .sort_values('count', ascending=False) \
    .reset_index()
# print(client_succ_operations)
sns.displot(client_succ_operations['count'], kde=True)
```

```python
computer_operations = user_data \
    .query('success == True and platform == "computer"') \
    .groupby(['platform', 'age']) \
    .agg({'platform': 'count'}) \
    .rename(columns={'platform': 'count'}) \
    .sort_values('count', ascending=False) \
    .reset_index()
computer_operations
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='age', data=user_data.query('success == True and platform == "computer"'))
```
