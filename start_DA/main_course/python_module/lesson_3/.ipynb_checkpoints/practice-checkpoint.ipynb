```python
file_adress = '3_companies.csv'
def read_n_agg(file_adress):
    df = pd.read_csv(file_adress, sep=';')
    income = df \
        .groupby(['company']) \
        .aggregate({'income': 'mean'})
    return income
read_n_agg(file_adress)
```

```python
taxi = pd.read_csv('3_taxi_peru.csv', parse_dates=['start_at', 'end_at', 'arrived_at'], sep=';')
```

```python
source_sum = taxi['source'].shape[0]
for source_name in taxi['source'].unique():
    print('Платформа: ' + str(source_name) + ' Кол-во заказов: ' + str(taxi.query(f'source == "{source_name}"').shape[0]))
    print(f'Процент платформы от общего кол-ва заказов: {int((taxi.query(f'source == "{source_name}"').shape[0] / source_sum) * 100)}%')
```

```python
platform_counter = taxi \
    .groupby('source') \
    .agg({'source': 'count'})
platform_counter.rename(columns={'source': 'Count'}, inplace=True)
platform_counter = platform_counter \
    .reset_index() \
    .sort_values('Count', ascending=False)
platform_counter.rename(columns={'source': 'Platform'}, inplace=True)

sns.barplot(x='Platform', y='Count', hue='Platform', data=platform_counter)
```

```python
sns.countplot(taxi, x='driver_score', hue='driver_score')
```

```python
driver_score_counts = taxi \
    .groupby('driver_score') \
    .aggregate({'driver_score': 'count'}) \
    .div(taxi['driver_score'].value_counts().sum()) \
    .mul(100) \
    .round(2) \
    .rename(columns={'driver_score': 'percentage'}) \
    .reset_index()
driver_score_counts
```

```python
ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage')
sns.despine()
```

```python
rider_score_counts = taxi \
    .groupby('rider_score') \
    .aggregate({'rider_score': 'count'}) \
    .div(taxi['rider_score'].value_counts().sum()) \
    .mul(100) \
    .round(2) \
    .rename(columns={'rider_score': 'percentage'}) \
    .reset_index()
rider_score_counts
```

```python
ax = sns.barplot(x='rider_score', y='percentage', data=rider_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Rider score', ylabel='Percentage')
sns.despine()
```
