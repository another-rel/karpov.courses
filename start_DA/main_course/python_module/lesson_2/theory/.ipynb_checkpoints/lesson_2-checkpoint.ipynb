```python
df = df.rename(columns = {'Номер': 'number',
          'Дата создания': 'create_date',
          'Дата оплаты': 'payment_date',
          'Title': 'title',
          'Статус': 'status',
          'Заработано': 'paycheck',
          'Город': 'city',
          'Платежная система': 'payment_system'})
```

```python
all_money = df.paycheck.sum()
all_money
```

```python
paycheck_by_city = df.groupby(['title', 'city'], as_index=False) \
    .aggregate({'paycheck': 'sum'}) \
    .sort_values('paycheck', ascending=False)
paycheck_by_city.to_csv('paycheck_by_city.csv', index=False)
```

```python
df.query('status == "Завершен"') \
    .groupby(['title'], as_index=False) \
    .aggregate({'paycheck': 'sum', 'number': 'count'}) \
    .sort_values('paycheck', ascending=False)
```
