```python
import pandas as pd
```

```python
bookings = pd.read_csv('2_bookings.csv', sep=';')
bookings_head = bookings.head(7)
bookings_head
```

```python
bookings.columns
```

```python
bookings.columns = bookings.columns.str.lower()
bookings.columns = bookings.columns.str.replace(' ', '_')
bookings.columns
```

```python
reservations_by_country = bookings \
    .query('is_canceled != 1') \
    .groupby('country', as_index=False) \
    .aggregate({'reserved_room_type': 'count'}) \
    .sort_values('reserved_room_type', ascending=False)
reservations_by_country.head(5)
```

```python
city_hotel_nights_mean = bookings \
    .query('hotel == "City Hotel"') \
    .groupby('hotel', as_index=False) \
    .aggregate({'stays_total_nights': 'mean'})
city_hotel_nights_mean['stays_total_nights'] = city_hotel_nights_mean['stays_total_nights'].round(2)
city_hotel_nights_mean
```

```python
resort_hotel_nights_mean = bookings \
    .query('hotel == "Resort Hotel"') \
    .groupby('hotel', as_index=False) \
    .aggregate({'stays_total_nights': 'mean'})
resort_hotel_nights_mean['stays_total_nights'] = resort_hotel_nights_mean['stays_total_nights'].round(2)
resort_hotel_nights_mean
```

```python
diff_room_type = bookings \
    .query('reserved_room_type != assigned_room_type') \
    .count()
diff_room_type
```

```python
arrival_mon_by_year_2016 = bookings \
    .query('arrival_date_year == 2016') \
    .groupby('arrival_date_month', as_index=False) \
    .aggregate({'reserved_room_type': 'count'}) \
    .sort_values('reserved_room_type', ascending=False)
arrival_mon_by_year_2016
```

```python
arrival_mon_by_year_2017 = bookings \
    .query('arrival_date_year == 2017') \
    .groupby('arrival_date_month', as_index=False) \
    .aggregate({'reserved_room_type': 'count'}) \
    .sort_values('reserved_room_type', ascending=False)
arrival_mon_by_year_2017
```

```python
cancel_rate_by_month = bookings \
    .query('hotel == "Resort Hotel" and is_canceled == 1') \
    .groupby(['arrival_date_year'], as_index=False)['arrival_date_month'].value_counts() \
    .sort_values(['arrival_date_year', 'count'], ascending=False)
cancel_rate_by_month
```

```python
print(bookings['adults'].mean())
print(bookings['children'].mean())
print(bookings['babies'].mean())
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
total_kids_by_hotel = bookings \
    .groupby('hotel') \
    .aggregate({'total_kids': 'mean'}) \
    .round(2)
total_kids_by_hotel
```

```python
bookings['has_kids'] = bookings['total_kids'] != 0
print(bookings['has_kids'])
```

```python
churn_rate_with_kids = bookings \
    .query('is_canceled == True and has_kids == True')
cr_withkids = churn_rate_with_kids.shape[0]

churn_rate_without_kids = bookings \
    .query('is_canceled == True and has_kids == False')
cr_nokids = churn_rate_without_kids.shape[0]

churn_rate_all_kids = bookings \
    .query('has_kids == True')
cr_all = churn_rate_all.shape[0]

churn_rate_all_no_kids = bookings \
    .query('has_kids == False')
cr_all_no_kids = churn_rate_all_no_kids.shape[0]

def calculate_percent(series1, series2):
    print(round((series1 / series2) * 100, 2))

print(cr_withkids)
print(cr_nokids)
print(cr_all)
calculate_percent(cr_withkids, cr_all)
calculate_percent(cr_nokids, cr_all_no_kids)
```
