```python
taxi = taxi.rename(columns = {'pcp 01': 'pcp_01', 'pcp 06': 'pcp_06', 'pcp 24': 'pcp_24'})
taxi
```

```python
taxi.columns
```

```python
taxi.query('borough == "Brooklyn"').shape
```

```python
taxi.aggregate({'pickups': 'sum'})
```

```python
min_pickups = taxi.groupby('borough') \
    .aggregate({'pickups': 'sum'}) \
    .sort_values('pickups', ascending=False) \
    .idxmin()
min_pickups
```

```python
borough_by_pickups = taxi \
    .groupby(['borough', 'hday'], as_index=False) \
    .aggregate({'pickups': 'mean'}) \
    .sort_values('pickups', ascending=False)
borough_by_pickups
```

```python
pickups_by_mon_bor = taxi \
    .groupby(['pickup_month', 'borough'], as_index=False) \
    .aggregate({'pickups': 'sum'}) \
    .sort_values(by = ['pickups'], ascending=False)
print(pickups_by_mon_bor)
```

```python
def temp_to_celcius(series):
    return (((series - 32) * 5) / 9)

taxi['temp_C'] = temp_to_celcius(taxi['temp'])
taxi
```
